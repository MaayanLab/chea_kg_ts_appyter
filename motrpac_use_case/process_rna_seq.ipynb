```python
for file in glob.glob("./raw_counts/rat/*.txt"):
    tissue = re.sub(r't[0-9]*-', '', file.split("_")[4])
    annotations = "../../motrpac_rna-seq_sample_annotations.csv"

    os.makedirs(f"out/rat/processed_rnaseq/{tissue}", exist_ok=True)
    os.chdir(f"out/rat/processed_rnaseq/{tissue}")
    print(clean_and_name_raw_data(file.replace("./", "../../"), tissue))
    rna_seq_data = f"{tissue}_rna-seq_named.txt"
    
    generate_metadata(rna_seq_data, annotations, tissue)
    os.chdir(f"../../")
```

## Human data preprocessing

```python
all_counts = dict()
for file in glob.glob("./raw_counts/human/*.txt"):
    tissue = "-".join(file.split("/")[-1].split("_")[2].split("-", 1)[1:])
    raw_data = pd.read_csv(file, sep="\t", index_col=0)
    raw_data.index = raw_data.index.map(lambda x: x.split(".")[0])
    ensembl_ids = raw_data.index.tolist()
    mg = mygene.MyGeneInfo()
    results = mg.querymany(ensembl_ids, scopes='ensembl.gene', fields='symbol', species='rat')

    id_to_symbol = {}
    for item in results:
        if 'symbol' in item:
            id_to_symbol[item['query']] = item.get('symbol', item['query']).upper()
        else:
            id_to_symbol[item['query']] = item['query']

    raw_data.index = raw_data.index.map(lambda x: id_to_symbol.get(x, x))
    raw_data = raw_data.drop_duplicates(subset='gene_id', keep='first')
    all_counts[tissue] = raw_data
    print(tissue)
    print(raw_data.head())
```
