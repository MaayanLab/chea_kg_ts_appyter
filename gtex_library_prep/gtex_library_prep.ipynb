This notebook is used to create the various tissue-specific GMT files for the ChEA-KG-TS appyter examples.

```python
import pandas as pd
from collections import defaultdict
```

```python
with open("./data/GTEx_Aging_Signatures_2021.txt", 'r') as f:
    gmt = {line.split("\t")[0]:line.split("\t")[2:] for line in f}
```

```python
terms_by_tissue = defaultdict(list)
for term in gmt:
    tissue = term.split(" ")[1]
    terms_by_tissue[tissue].append(term)
```

```python
terms_by_tissue = {k: sorted(v) for k,v in terms_by_tissue.items()}
terms_by_tissue['Uterus']
```

```python
for tissue, terms in terms_by_tissue.items():
    sub_gmt = {}
    sep = '\t'
    for term in terms:
        formatted_term = term.replace(f"GTEx {tissue}", "").replace("Up", "up genes").replace("Down", "down genes")
        sub_gmt[formatted_term] = gmt[term]
    with open(f"./out/gtex_aging_{tissue}.gmt", "w") as f:
        for term, set in sub_gmt.items():
            f.write(f'{term}\t\t{sep.join(set)}')
```
